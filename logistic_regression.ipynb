# Logistic Regression

### Variables to test: <br>-Total day minutes  <br>-Total eve minutes<br>-Total intl minutes <br>-Customer services calls

```python
col = pd.DataFrame(data[['Total day minutes' , 'Total eve minutes' , 'Total intl minutes', "Customer service calls"]])
col
```

### Response variable: Churn

```python
churn = pd.DataFrame(data['Churn'])
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import f1_score
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()     # Standardize data before Logistic Regression
```

```python
# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(scale.fit_transform(col), churn, test_size = 0.25, random_state= 42)
```

```python
logreg = LogisticRegression()
logreg.fit(col,churn)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix


class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sb.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
target_names = ['No Churn', 'Churn']
print(classification_report(y_test, y_pred, target_names=target_names))
```

### AOC ROC 

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
